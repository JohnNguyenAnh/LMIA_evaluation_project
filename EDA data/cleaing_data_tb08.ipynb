```python
import pandas as pd
import os
```

```python
#import table file
df = pd.read_csv(r'../raw_data/Table08_TFW_Positive_LMIAs_Caregivers_NOC_Province_2016_2023.csv', encoding='ISO-8859-1',skiprows=1)
df.tail(20)
```

```python
df
```

```python
df = df.iloc[:-8]
```

```python
df.tail(10)
```

```python
df.head(10)
```

```python
df.rename(columns={
    'Unnamed: 0':'NOC Code and Occupational Group',
    'Unnamed: 1' : '2016',
    'Unnamed: 2' : '2017',	
    'Unnamed: 3' : '2018',	
    'Unnamed: 4' : '2019',	
    'Unnamed: 5' : '2020',	
    'Unnamed: 6' : '2021',	
    'Unnamed: 7' : '2022',	
    'Unnamed: 8' : '2023'
}, inplace=True)
df
```

```python
#  Remove rows that contain "NOC Code and Occupational Group" adn Nan value 
df = df[~df['NOC Code and Occupational Group'].str.contains('NOC', na=False)]
#Create a new column for province/region name
df['province_region'] = df['NOC Code and Occupational Group'].where(df['2016'].isna() & ~df['NOC Code and Occupational Group'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NOC Code and Occupational Group'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
df = df.dropna(how='any')
```

```python
df
```

```python
# Split the column 'NOC Code and Occupational Group' 
df[['NOC Code', 'Occupational Group']] = df['NOC Code and Occupational Group'].str.extract(r'(\w+) - (.*)')
#drop the 'NOC Code and Occupational Group' column
df = df.drop(columns=['NOC Code and Occupational Group'])
```

```python
df
```

```python
#Define the desired column order
desired_order = ['province_region',
    'NOC Code',
    'Occupational Group',
    '2016',
    '2017',
    '2018',
    '2019',
    '2020', 
    '2021', 
    '2022', 
    '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```

```python
df
```

```python
df = df.reset_index(drop=True)
```
