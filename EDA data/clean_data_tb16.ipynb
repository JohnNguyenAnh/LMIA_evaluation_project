```python
import pandas as pd
import os
```

```python
df = pd.read_csv(r'../raw_data/Table16_Positive_LMIAs_By_NOC_SkillLevel_Province_2016_2023.csv', encoding='ISO-8859-1')
df.tail(20)
```

```python
df = df.iloc[:-7]
```

```python
df.tail()
```

```python
df.head(20)
```

```python
# Define new column names in the desired order
new_column_names = ['Occupational Skill Level', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names
```

```python
df.head(25)
```

```python
#  Remove rows that contain "Occupational Skill Level" adn Nan value (since these are summary rows)
df = df[~df['Occupational Skill Level'].str.contains('Occupational', na=True)]
```

```python
df
```

```python
#Create a new column for province/region name
df['province_region'] = df['Occupational Skill Level'].where(df['2016'].isna() & ~df['Occupational Skill Level'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Occupational Skill Level'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
df = df.dropna(how='any')
```

```python
df.head(20)
```

```python
#Create new columns for skill level  and Occupational code  seperatelly
df[['Skill Level','Occupation']] = df['Occupational Skill Level'].str.extract(r'(\w+) - (.*)')
# Drop the original column after splitting
df = df.drop(columns=['Occupational Skill Level'])
```

```python
df.head(20)
```

```python
# Define the desired column order
desired_order = ['province_region', 'Occupation','Skill Level', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```

```python
df
```

```python
df = df.reset_index(drop=True)
```
