```python
import pandas as pd
import os
```

```python
#import table 4 file
df = pd.read_csv(r'../raw_data/Table06_TFW_Positive_LMIAs_By_NAICS_Province_2016_2023.csv', encoding='ISO-8859-1')
df
```

```python
df = df.iloc[:-7]
```

```python
df.head(20)
```

```python
#rename the 1st column
df.rename(columns={
    'Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by North American Industry Classification System - 2002 (NAICS) and province/territory': 'NAICS Codes and Industry Sectors',
    'Unnamed: 1' : '2016',
    'Unnamed: 2' : '2017',	
    'Unnamed: 3' : '2018',	
    'Unnamed: 4' : '2019',	
    'Unnamed: 5' : '2020',	
    'Unnamed: 6' : '2021',	
    'Unnamed: 7' : '2022',	
    'Unnamed: 8' : '2023'
    }, inplace= True)
```

```python
df.head(30)
```

```python
df = df[~df['NAICS Codes and Industry Sectors'].str.contains('NAICS', na=True)]
```

```python
df.head(30)
```

```python
#Create a new column for province/region name
df['Province_Region'] = df['NAICS Codes and Industry Sectors'].where(df['2016'].isna() & ~df['NAICS Codes and Industry Sectors'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['Province_Region'] = df['Province_Region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Province_Region'].str.contains('Total', na=False)]
#drop unessesary columns
df = df.dropna(how='any')
```

```python
df.head(30)
```

```python
#Create new columns for skill level  and Occupational code  seperatelly
df[['NAICS_Codes ','Occupation']] = df['NAICS Codes and Industry Sectors'].str.extract(r'(\d+(?:-\d+)?)\s+(.*)')

# Create a new column called 'Occupation' and initialize it with NaN
df['Industry_Sectors'] = None

# Loop through the DataFrame in reverse order to propagate the occupation names upwards
current_occupation = None
for i in reversed(df.index):
    if pd.notna(df.loc[i, 'NAICS Codes and Industry Sectors']) and df.loc[i, 'NAICS Codes and Industry Sectors'].endswith('- Total'):
        # If the row contains an occupation name ending with '- Total', update current_occupation without the suffix
        current_occupation = df.loc[i, 'NAICS Codes and Industry Sectors'].replace(' - Total', '')
    # Assign the current occupation to the new 'Occupation' column for each row
    df.loc[i, 'Industry_Sectors'] = current_occupation
```

```python
#Drop rows where Nan value
df = df.dropna(how='any')

#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NAICS Codes and Industry Sectors'].str.contains('Total', na=False)]
```

```python
df.head(30)
```

```python
#drop no Nan and unneccessary columns
df = df.dropna(how='any')
df = df.drop(columns=['NAICS Codes and Industry Sectors'])
#reset index
df = df.reset_index(drop=True)
```
