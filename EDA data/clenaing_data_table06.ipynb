```python
import pandas as pd
import os
```

```python
#import table 4 file
df = pd.read_csv(r'../raw_data/Table06_TFW_Positive_LMIAs_By_NAICS_Province_2016_2023.csv', encoding='ISO-8859-1')
df
```

```python
df = df.iloc[:-7]
```

```python
df.head(20)
```

```python
#rename the 1st column
df.rename(columns={
    'Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by North American Industry Classification System - 2002 (NAICS) and province/territory': 'NAICS Codes and Industry Sectors',
    'Unnamed: 1' : '2016',
    'Unnamed: 2' : '2017',	
    'Unnamed: 3' : '2018',	
    'Unnamed: 4' : '2019',	
    'Unnamed: 5' : '2020',	
    'Unnamed: 6' : '2021',	
    'Unnamed: 7' : '2022',	
    'Unnamed: 8' : '2023'
    }, inplace= True)
```

```python
df.head(30)
```

```python
df =df.dropna(how='all')
df = df[~df['NAICS Codes and Industry Sectors'].str.contains('NAICS', na=True)]
```

```python
df
```

```python
#Create a new column for province/region name
df['Province_Region'] = df['NAICS Codes and Industry Sectors'].where(df['2016'].isna() & ~df['NAICS Codes and Industry Sectors'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['Province_Region'] = df['Province_Region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NAICS Codes and Industry Sectors'].str.contains('Total', na=False)]
```

```python
df
```

```python
#Drop rows where Nan value
df = df.dropna(how='any')
```

```python
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NAICS Codes and Industry Sectors'].str.contains('Total', na=False)]
```

```python
df
```

```python
# Extract NAICS Codes (numbers) and Industry Sectors (words)
df["NAICS_Codes"] = df["NAICS Codes and Industry Sectors"].str.extract(r"^(\d+-?\d*)")
df["Industry_Sectors"] = df["NAICS Codes and Industry Sectors"].str.replace(r"^\d+-?\d*\s*", "", regex=True)

# Drop the original column (optional)
df = df.drop("NAICS Codes and Industry Sectors", axis=1)
```

```python
df
```

```python
desired_order = ['Province_Region','NAICS_Codes','Industry_Sectors', '2016',	'2017',	'2018',	'2019','2020','2021','2022','2023']
df = df[desired_order]
```

```python
df
```

```python
df =df.reset_index(drop= True)
```
