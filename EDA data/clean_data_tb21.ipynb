```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table21_Unique_Employers_Positive_LMIAs_By_NAICS_Province_2016_2023.csv', encoding='ISO-8859-1')
df.tail(20)
```

```python
df = df.iloc[:-7]
df.tail()
```

```python
df.head(15)
```

```python
# Define new column names in the desired order
new_column_names = ['NAICS_Codes_Industry_Sectors', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names

#  Remove rows that contain "Occupational Skill Level" adn Nan value (since these are summary rows)
df = df[~df['NAICS_Codes_Industry_Sectors'].str.contains('NAICS', na=True)]
```

```python
df.head(20)
```

```python
#Create a new column for province/region name
df['province_region'] = df['NAICS_Codes_Industry_Sectors'].where(df['2016'].isna() & ~df['NAICS_Codes_Industry_Sectors'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NAICS_Codes_Industry_Sectors'].str.contains('Total', na=False)]
```

```python
df.sample(20)
```

```python
df = df.dropna(how='any')
```

```python
df.head(20)
```

```python
# Extract the skill level (numbers) and definition (industry sector)
df[['NAICS_Codes', 'Industry_Sector']] = df['NAICS_Codes_Industry_Sectors'].str.extract(r'(\d{2,3}-\d{2,3}|\d+) (.*)')

# Drop the original column after splitting
df = df.drop(columns=['NAICS_Codes_Industry_Sectors'])
```

```python
df.head(20)
```

```python
# Define the desired column order
desired_order = ['province_region', 'NAICS_Codes','Industry_Sector', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```

```python
df = df.reset_index(drop=True)
```
