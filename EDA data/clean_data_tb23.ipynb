```python
import pandas as pd
import os
```

```python
# Use a raw string literal for the file path
file_path = r'C:\Users\Johnn\Desktop\Data Engineer project\LMIA_evaluation_project\raw_data\Table23_TFW_Positive_LMIAs_By_EconomicRegion_NOC_SkillLevel_Province_2016_2023.xlsx'

# Load the Excel file using the openpyxl engine
df = pd.read_excel(file_path, engine='openpyxl',skiprows=1)

# Display the first few rows of the data to verify it loaded correctly
df.head()
```

```python
# Convert all numeric columns to the nullable integer type to handle missing values properly
for column in df.select_dtypes(include='float').columns:
    df[column] = df[column].astype('Int64')  # Use 'Int64' to support missing values

# Display the first few rows to verify the conversion
df.head(20)
```

```python
# Define new column names in the desired order
new_column_names = ['province_region','Occupational Skill Level', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names

```
