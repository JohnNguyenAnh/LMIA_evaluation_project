```python
import pandas as pd
import os
```

```python
#import table file
df = pd.read_csv(r'../raw_data/Table03_TFW_Positive_LMIAs_By_EconomicRegion_Province_2016_2023.csv', encoding='ISO-8859-1')
df.head()
```

```python
df = df.iloc[:-11]
df
```

```python
df.head(20)
```

```python
#Create a new column for province/region name
# Identify rows where 'Unnamed: 1' is NaN (indicating region rows) and exclude rows containing 'Total'
df['province_region'] = df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].where(df['Unnamed: 1'].isna() & ~df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
# The rows after region names have NaN values for the year columns
df_cleaned = df.dropna(subset=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8'], how='all')

# Find the row where "Economic Region" is the column value and set that as the new header
new_headers = df_cleaned[df_cleaned['Economic Region'].str.contains('Economic Region', na=False)].iloc[0]

# Rename the columns based on the new header row
df_cleaned.columns = new_headers

# Remove all rows where "Economic Region" is present since we've used them as headers
df_cleaned = df_cleaned[~df_cleaned['Economic Region'].str.contains('Economic Region', na=False)]

#Extract region codes into a new column
# Extract numeric code
df_cleaned['region_code'] = df_cleaned['Economic Region'].str.extract(r'(\d+)') 
# Remove code from region name 
df_cleaned['Economic Region'] = df_cleaned['Economic Region'].str.replace(r'^\d+\s+-\s+', '')  
#Rename the province_region column
df_cleaned.rename(columns={'Newfoundland and Labrador': 'province_region'}, inplace=True)
```
