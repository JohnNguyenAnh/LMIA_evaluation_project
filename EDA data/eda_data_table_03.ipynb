```python
import pandas as pd
import os
```

```python
#import table file
df = pd.read_csv(r'../raw_data/Table03_TFW_Positive_LMIAs_By_EconomicRegion_Province_2016_2023.csv', encoding='ISO-8859-1')
df.head()
```

```python
df = df.iloc[:-11]
df
```

```python
df.head(20)
```

```python
#Create a new column for province/region name
# Identify rows where 'Unnamed: 1' is NaN (indicating region rows) and exclude rows containing 'Total'
df['province_region'] = df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].where(df['Unnamed: 1'].isna() & ~df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by Statistics Canada economic region and province/territory'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
# Define new column names in the desired order
new_column_names = ['Economic_Region', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', 'Province_Territory']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names
```

```python
df.head(20)
```

```python
df = df.dropna(how='any')
df.head(15)
```

```python
df = df[~df['Economic_Region'].str.contains('Economic Region', case=False, na=False)]
```

```python
df
```

```python
df = df[~df['Economic_Region'].str.contains('Other', case=False, na=False)]
```

```python
df
```

```python
#Create new columns for 'Region_Code','Region_Name'  seperatelly
df[['Region_Code','Region_Name']] = df['Economic_Region'].str.extract(r'(\w+) - (.*)')
# Drop the original column after splitting
df = df.drop(columns=['Economic_Region'])
```
