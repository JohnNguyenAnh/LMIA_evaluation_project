```python
# Remove rows where 'Province' is 'Total'
df = df[df['Province/Territory'] != 'Total'].reset_index(drop=True)
```

```python
df[['CSD_Code', 'CSD_Name']] = df['CSD'].str.split('-', n=1, expand=True)
df = df.drop('CSD', axis=1)
```
