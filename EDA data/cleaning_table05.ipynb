```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table05_TFW_Positive_LMIAs_By_NOC_SkillType_Province_2016_2023.csv', encoding='ISO-8859-1',skiprows=1)
df.tail(20)
```

```python
df = df.iloc[:-7]
```

```python
df.head(20)
```

```python
#rename columns

df.rename(columns={
    'Unnamed: 0':'Skill Type and Description',	
    'Unnamed: 1':'2016'	,
    'Unnamed: 2':'2017'	,
    'Unnamed: 3':'2018'	,
    'Unnamed: 4':'2019'	,
    'Unnamed: 5':'2020'	,
    'Unnamed: 6':'2021'	,
    'Unnamed: 7':'2022'	,
    'Unnamed: 8':'2023'
    }, inplace=True)
```

```python
df.head(20)
```

```python
#Create a new column for province/region name
df['province_region'] = df['Skill Type and Description'].where(df['2016'].isna() & ~df['Skill Type and Description'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Skill Type and Description'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
# The rows after region names have NaN values for the year columns
df = df.dropna(subset=['2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'], how='all')
df.head(20)
```

```python
df = df[df['Skill Type and Description'] != 'Skill Type and Description']
df.head(20)
```

```python
# Split the column 'Management Occupations and Skill Level (A - D)' into 'skill_level' and 'occupation'
df[['Skill Type', 'Description']] = df['Skill Type and Description'].str.extract(r'(\w+) - (.*)')

# Drop the original column after splitting
df = df.drop(columns=['Skill Type and Description'])
```

```python
df.head(10)
```

```python
# Define the desired column order
desired_order = ['province_region', 'Description','Skill Type', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```

```python
df.head(10)
```

```python
# Reset the index of the DataFrame
df = df.reset_index(drop=True)
```

```python
df.head(10)
```
