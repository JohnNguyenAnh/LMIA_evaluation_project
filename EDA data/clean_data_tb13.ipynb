```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table13_TFW_Positive_LMIAs_By_Stream_Province_2016_2023.csv', encoding='ISO-8859-1',skiprows=1)
df.tail(20)
```

```python
df = df.iloc[:-8]
```

```python
df.tail()
```

```python
df.head(20)
```

```python
#remove row contain "stream"
df['Unnamed: 0'] = df['Unnamed: 0'].where(~df['Unnamed: 0'].str.contains('Stream', na=False))
```

```python
df
```

```python
#Create a new column for province/region name
df['province_region'] = df['Unnamed: 0'].where(df['Unnamed: 1'].isna() & ~df['Unnamed: 0'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Unnamed: 0'].str.contains('Total', na=False)]
```

```python
df
```

```python
df = df.dropna(how='any')
```

```python
df.head(20)
```

```python
df.rename(columns={
    'Unnamed: 0' :	'Stream',
    'Unnamed: 1' :	'2016',
    'Unnamed: 2' :	'2017',
    'Unnamed: 3' :	'2018',
    'Unnamed: 4' :	'2019',
    'Unnamed: 5' :	'2020',
    'Unnamed: 6' :	'2021',
    'Unnamed: 7' :	'2022',
    'Unnamed: 8' :	'2023'
}, inplace=True)
```
