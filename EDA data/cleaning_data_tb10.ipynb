```python
import pandas as pd
import os
```

```python
#import table  file
df = pd.read_csv(r'../raw_data/Table10_TFW_Positive_LMIAs_Top_Countries_Residency_Province_2020_2023.csv', encoding='ISO-8859-1',skiprows=1)
df.tail(20)
```

```python
df = df.iloc[:-7]
df.tail()
```

```python
df = df.dropna(how='all')
```

```python
df.tail(30)
```

```python
df.head(15)
```

```python
# Identify rows with year information and set them to a new 'year' column
df['year'] = df['Unnamed: 0'].where(df['Unnamed: 0'].str.isnumeric())
df['year'] = df['year'].ffill()
```

```python
df.head()
```

```python
#Extract province/region and forward-fill
df['province_region'] = df['Unnamed: 0'].fillna('').where(
     df['Unnamed: 1'].isna() &
    ~df['Unnamed: 0'].fillna('').str.isnumeric() & 
    ~df['Unnamed: 0'].fillna('').str.contains('Country of Citizenship', na=False) & 
    ~df['Unnamed: 0'].fillna('').str.contains('Rank', na=False)
)
```

```python
df.head(20)
```

```python
# Forward fill the province names
df['province_region'] = df['province_region'].ffill()
```

```python
df.sample(15)
```

```python
df = df.dropna(how='any')
```

```python
df.sample(15)
```

```python
#rename columns

df.rename(columns={
    'Unnamed: 0':'Country of Citizenship',	
    'Unnamed: 1':'Positions'	,
    'Unnamed: 2':'Rank'
    
    }, inplace=True)
```

```python
df
```

```python
# Drop rows where 'province_region' contains all of the following: "Country of Citizenship", "Positions", and "Rank"
df = df[~df['Country of Citizenship'].str.contains('Country of Citizenship|Positions|Rank', na=False)]
```

```python
df.head(15)
```

```python
df = df.reset_index(drop=True)
```
