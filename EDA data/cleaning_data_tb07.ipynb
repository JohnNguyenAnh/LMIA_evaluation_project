```python
import pandas as pd
import os
```

```python
#import table 4 file
df = pd.read_csv(r'../raw_data/Table07_TFW_Positive_LMIAs_Agriculture_Province_2016_2023.csv', encoding='ISO-8859-1')
df.tail(15)
```

```python
df.head(50)
```

```python
df = df.iloc[:-8]
```

```python
df.tail()
```

```python
df.rename(columns={
    'Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by National Occupational Classification - 2011 (NOC), and province/territory':'NOC Code and Occupational Group',
    'Unnamed: 1' : '2016',
    'Unnamed: 2' : '2017',	
    'Unnamed: 3' : '2018',	
    'Unnamed: 4' : '2019',	
    'Unnamed: 5' : '2020',	
    'Unnamed: 6' : '2021',	
    'Unnamed: 7' : '2022',	
    'Unnamed: 8' : '2023'
}, inplace=True)
df
```

```python
#  Remove rows that contain "NOC Code and Occupational Group" adn Nan value 
df = df[~df['NOC Code and Occupational Group'].str.contains('NOC Code and Occupational Group', na=False)]
```

```python
#Create a new column for province/region name
df['province_region'] = df['NOC Code and Occupational Group'].where(df['2016'].isna() & ~df['NOC Code and Occupational Group'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['NOC Code and Occupational Group'].str.contains('Total', na=False)]
```

```python
df
```

```python
# The rows after region names have NaN values for the year columns
df = df.dropna(subset=['NOC Code and Occupational Group','2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'], how='any')
df.head(20)
```

```python
df.sample(10)
```

```python
#Create new columns for NOC code and Occupation group seperatelly
df[['NOC code','Occupational Group']] = df['NOC Code and Occupational Group'].str.extract(r'(\w+) - (.*)')
```

```python
df.head()
```

```python
# Drop the original column after splitting
df = df.drop(columns=['NOC Code and Occupational Group'])
```

```python
df.head(15)
```

```python
# reset the index of data
df = df.reset_index(drop=True)
```

```python
#Define the desired column order
desired_order = ['province_region',
    'NOC code',
    'Occupational Group',
    '2016',
    '2017',
    '2018',
    '2019',
    '2020', 
    '2021', 
    '2022', 
    '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```
