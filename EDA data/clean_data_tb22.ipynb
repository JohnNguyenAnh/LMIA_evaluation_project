```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table22_TFW_Positive_LMIAs_By_EconomicRegion_Stream_Province_2016_2023.csv', encoding='ISO-8859-1')
df.tail(15)
```

```python
df = df.iloc[:-8]
df.tail()
```

```python
df.head(15)
```

```python
# Define new column names in the desired order
new_column_names = ['Economic_Region','Program Stream', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names

```

```python
df.sample(15)
```

```python
#Create a new column for province/region name
df['province_region'] = df['Economic_Region'].where(df['2016'].isna() & ~df['Economic_Region'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Economic_Region'].str.contains('Total', na=False)]
```

```python
df.head(301)
```

```python
#Create new columns for skill level  and Occupational code  seperatelly
df[['Region_code','Region_Name']] = df['Economic_Region'].str.extract(r'(\w+) - (.*)')
# Drop the original column after splitting
df = df.drop(columns=['Economic_Region'])
#drop nan value
df = df.dropna(how='any')
```

```python
df.head(20)
```

```python
df = df.reset_index(drop=True)
```
