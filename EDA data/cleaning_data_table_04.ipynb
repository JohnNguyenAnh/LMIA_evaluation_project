```python
import pandas as pd
import os
```

```python
#import table 4 file
df = pd.read_csv(r'../raw_data/Table04_TFW_Positive_LMIAs_By_NOC_SkillLevel_Province_2016_2023.csv', encoding='ISO-8859-1')
df.tail(20)

```

```python
df = df.iloc[:-7]
df.head(20)
```

```python
#Rename the first column
df.rename(columns={'Number of temporary foreign worker (TFW) positions on positive Labour Market Impact Assessments (LMIAs) by National Occupational Classification - 2011 (NOC) skill level and management group of occupations, and province/territory': 'province_region'}, inplace=True)
```

```python
df.head(20)
```

```python
# Identify the row that contains headers and use it as the new header
header_row_index = df[df['province_region'].str.contains('Management Occupations', na=False)].index[0]
new_headers = df.loc[header_row_index]
df.columns = new_headers
df.head(10)
```

```python
# Drop the header rows after setting them
df = df.drop(header_row_index)
df.head(20)
```

```python
#Create a new column for province/region name
# Identify rows where 'Unnamed: 1' is NaN (indicating region rows) and exclude rows containing 'Total'
df['province_region'] = df['Management Occupations and Skill Level (A - D)'].where(df['2016'].isna() & ~df['Management Occupations and Skill Level (A - D)'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Management Occupations and Skill Level (A - D)'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
# The rows after region names have NaN values for the year columns
df = df.dropna(subset=['2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'], how='all')
df.head(20)
```

```python
df = df[df['Management Occupations and Skill Level (A - D)'] != 'Management Occupations and Skill Level (A - D)']
df.head(20)
```

```python
# Split the column 'Management Occupations and Skill Level (A - D)' into 'skill_level' and 'occupation'
df[['skill_level', 'occupation']] = df['Management Occupations and Skill Level (A - D)'].str.extract(r'(\w+) - (.*)')

# Drop the original column after splitting
df = df.drop(columns=['Management Occupations and Skill Level (A - D)'])
```

```python
df.head(5)
```

```python
# Define the desired column order
desired_order = ['province_region', 'occupation','skill_level', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
# Reorder the DataFrame using the desired column order
df = df[desired_order]
```
