```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table24_TFW_Positive_LMIAs_By_EconomicRegion_NOC_SkillType_Province_2016_2023.csv', encoding='ISO-8859-1',skiprows=1)
df.tail(20)
```

```python
df = df.iloc[:-9]
df.tail()
```

```python
df.head()
```

```python
# Define new column names in the desired order
new_column_names = ['Province_Territory','Economic_Region','NOC_2011_Occupational_Skill_Type', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Rename the columns by mapping the old names to the new names in order
df.columns = new_column_names

```

```python
df.head()
```

```python
#Create new columns for 'Region_Code','Region_Name'  seperatelly
df[['Region_Code','Region_Name']] = df['Economic_Region'].str.extract(r'(\w+) - (.*)')
# Drop the original column after splitting
df = df.drop(columns=['Economic_Region'])
#Create new columns for NOC_2011_Occupational_Skill_Type  seperatelly
df[['NOC_Code','Occupational_Skill_Type']] = df['NOC_2011_Occupational_Skill_Type'].str.extract(r'(\w+) - (.*)')
# Drop the original column after splitting
df = df.drop(columns=['NOC_2011_Occupational_Skill_Type'])
```

```python
df
```

```python
df = df.dropna(how='any')
```

```python
df
```

```python
df = df.reset_index(drop=True)
```
