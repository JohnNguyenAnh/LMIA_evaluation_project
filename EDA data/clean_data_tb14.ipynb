```python
import pandas as pd
import os
```

```python
#import table 5 file
df = pd.read_csv(r'../raw_data/Table14_Positive_LMIAs_By_Stream_Province_2016_2023.csv', encoding='ISO-8859-1') #,skiprows=1 
df.tail(20)
```

```python
df = df.iloc[:-6]
```

```python
df.tail(10)
```

```python
df.head(10)
```

```python
#rename columns

df.rename(columns={
    'Number of Positive Labour Market Impact Assessments (LMIAs) by Program Stream and Province / Territory':'Program Stream',	
    'Unnamed: 1':'2016'	,
    'Unnamed: 2':'2017'	,
    'Unnamed: 3':'2018'	,
    'Unnamed: 4':'2019'	,
    'Unnamed: 5':'2020'	,
    'Unnamed: 6':'2021'	,
    'Unnamed: 7':'2022'	,
    'Unnamed: 8':'2023'
    }, inplace=True)
```

```python
df
```

```python
#remove "program stream" row
df = df[~df['Program Stream'].str.contains('Program Stream', na=False)]
```

```python
df.head(20)
```

```python
#Create a new column for province/region name
df['province_region'] = df['Program Stream'].where(df['2016'].isna() & ~df['Program Stream'].str.contains('Total', na=False))
#  Forward-fill the region name for all rows until a new region is found
df['province_region'] = df['province_region'].ffill()
#  Remove rows that contain "Total" adn Nan value (since these are summary rows)
df = df[~df['Program Stream'].str.contains('Total', na=False)]
```

```python
df.head(20)
```

```python
df = df.dropna(how="any")
```

```python
df.head(20)
```

```python
# Reset the index of the DataFrame
df = df.reset_index(drop=True)
```

```python
df.head(20)
```
